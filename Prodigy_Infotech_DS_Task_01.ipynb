Task: Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('/content/drive/MyDrive/heart_2020_cleaned.csv')
```

```python
sns.countplot(x='Sex',hue='HeartDisease',palette='Spectral',data=data)
plt.xlabel('Gender')
plt.ylabel('Population')
```

```python
sns.countplot(x='Smoking',palette='Spectral',data=data)
plt.xlabel('Smoking')
plt.ylabel('Population')
```

```python
categorical=data.select_dtypes(include='object').columns
for i in categorical:
  sns.barplot(x=data[i],y='BMI',data=data)
  plt.show()
```

```python
numerical=data.select_dtypes(include='float64').columns
for i in numerical:
  sns.histplot(x=data[i],data=data,bins=25,kde=True,color='orange')
  plt.show()
```
